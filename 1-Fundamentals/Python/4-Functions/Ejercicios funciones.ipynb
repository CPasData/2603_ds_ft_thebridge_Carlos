## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dias(numero): # dias es una variable y numero tambien. Además, numero es un parámetro de entrada, y el valor que le paso es el argumento.
    mi_dict = {1:"Lunes", 2:"Martes", 3:"Miercoles", 4:"Jueves", 5:"Viernes", 6:"Sabado", 7:"Domingo"}
    
    if numero in range(1,7):
        return mi_dict[numero]
    else:
        return ("Lo siento, debe ser un numero entre el 1 y el 7")

dias(6)
````

## Ejercicio 2
En el ejercicio 8 de bucles, creábamos una pirámide invertida, cuyo número de pisos venía determinado por un input del usuario. Crear una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide, es decir, elimina la sentencia input.

````python
def piramide(numero):
    lineas = len(range(numero))
    while lineas > 0:
        for x in range(1, lineas + 1):
            print(x, end=" ")
        print()
        lineas -= 1

piramide(5)
````

## Ejercicio 3
Escibe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero se  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar(num1, num2):
    if num1 == num2:
        return f"{num1} es igual a {num2}"
    elif num1 > num2:
        return f"{num1} es mayor a {num2}"
    else:
        return f"{num1} es menor a {num2}"
    
comparar(8, 254)
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea la letra a contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayuscula, como minúscula

````python
def contador(texto, letra):
    return texto.lower().count(letra.lower())

contador("Hola me llamo Carlos","m")
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string.

````python
def convertir_dicc(frase:str):
    return dict(enumerate(str(frase)))

convertir_dicc("Hola mundo")
````

````python
def letter_count_dic(texto):

    all_letters = {}

    for i in texto.lower():
        if i in all_letters:
            all_letters[i] = all_letters[i] + 1
        else:
            all_letters[i] = 1
    
    return all_letters

letter_count_dic("Hola, estamos en clase y es viernes")
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

````python
def modificar(lista, elemento = None,*, add = True, remove = False):
    if add:
        lista.append(elemento)
    if remove:
        if elemento in lista:
            lista.remove(elemento)
        
    return lista
    
modificar([1,2,3,4,5], 3, add=True, remove=True)
````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

````python
def fibonacci(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return (fibonacci(x-1) + fibonacci(x-2))
    
fibonacci(9)
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el area de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math

def area_cuadrado(base, altura):
    return base * altura

def area_triangulo(base, altura):
    return (base * altura) / 2

def area_circulo(radio):
    return (math.pi * (radio ** 2))
````

````python
circulos = (area_circulo (10)) * 2
triangulo = area_triangulo(3, 7)
ejer_1 = circulos + triangulo

print(ejer_1)

cuadrado = area_cuadrado(10,10)
circulos2 = area_circulo(4) + (area_circulo(6) * 2)
triangulos2 = area_triangulo(2, 4) * 5

ejer_2 = cuadrado + circulos2 + triangulos2

print(ejer_2)
````
