### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Nos piden hombres / total personas por ocupacion
```

```python
users.head()
```

```python
lista_n = []
for x in users["gender"]:
    if x == "M":
        lista_n.append(1)
    else:
        lista_n.append(0)

users["gender_n"] = lista_n
users
```

```python
users.groupby("occupation")["gender_n"].mean().sort_values(ascending=False)
```

```python
users.groupby("gender").mean(numeric_only=True)
```

```python
users["ratio"] = users.groupby("occupation")["gender_n"].transform("mean") * 100
users.sort_values(by="ratio", ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].describe()[["min", "max"]]
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])["age"].mean().to_frame()
```

### Step 8.  For each occupation present the percentage of women and men

```python
total = users.groupby(["occupation", "gender"]).size().unstack(fill_value=0)
p_mujer = (total.loc[:, "F"] / total.sum(axis=1)) * 100
p_hombre = (total.loc[:, "M"] / total.sum(axis=1)) * 100

total["Porcentaje_F"] = p_mujer
total["Porcentaje_M"] = p_hombre

total
```
